### nsw_experimental

Exemplo prático do livro "Causal Inference: The Mixtape" referente ao capítulo tratando sobre Pareamento de Pontuação de Propensão (*PSM - Propensity Score Matching*).

```python
# Importando os pacotes necessários
import numpy as np 
import pandas as pd 
import statsmodels.api as sm 
import statsmodels.formula.api as smf 
from itertools import combinations 
import plotnine as p
import ssl
import warnings
warnings.filterwarnings('ignore')
```

```python
# Criando uma função de importação/leitura dos dados
ssl._create_default_https_context = ssl._create_unverified_context
def read_data(file):
    return pd.read_stata("https://github.com/scunning1975/mixtape/raw/master/" + file)
```

```python
# Importando a base direto do github do autor usando a função criada acima
nsw_dw = read_data('nsw_mixtape.dta')
nsw_dw
```

```python
# Calculando o ATE ("Average Treatment Effect") do experimento da NSW
mean1 = nsw_dw[nsw_dw.treat==1].re78.mean() # Calculando a média dos ganhos dos tratados
mean0 = nsw_dw[nsw_dw.treat==0].re78.mean() # Calculando a média dos ganhos dos não-tratados (experimental)
ate = np.unique(mean1 - mean0)[0] # Tirando a diferença simples
print("O ATE experimental estimado é de ${:.2f}".format(ate)) # Printando nosso resultado de interesse
```

### nsw_pscore

```python
## Preparando os dados para a estimação logit 
nsw_dw = nsw_dw[nsw_dw.treat==1] # Mantendo apenas as observações do grupo de tratamento em nosso df primario
nsw_dw_cpscontrol = read_data('cps_mixtape.dta') # Importando a base alternativa para cálculo do PSM
nsw_dw_cpscontrol = pd.concat((nsw_dw_cpscontrol, nsw_dw)) # Concatenando as duas bases
nsw_dw_cpscontrol['agesq'] = np.power((nsw_dw_cpscontrol.age), 2) # Criando uma variável da idade ao quadrado
nsw_dw_cpscontrol['agecube'] = np.power((nsw_dw_cpscontrol.age), 3) # Criando uma variável da idade ao cubo
nsw_dw_cpscontrol['educsq'] = np.power((nsw_dw_cpscontrol.educ), 2) # Criando uma variável da escolaridade ao quadrado
# Criando dummies p/ indicar se as unidades estavam desempregadas ou não em 1974 e 1975
nsw_dw_cpscontrol['u74'], nsw_dw_cpscontrol['u75'] = 0, 0
nsw_dw_cpscontrol.loc[nsw_dw_cpscontrol.re74==0, 'u74'] = 1
nsw_dw_cpscontrol.loc[nsw_dw_cpscontrol.re75==0, 'u75'] = 1
# Criando uma variável de interação entre escolaridade e a renda de 1974
nsw_dw_cpscontrol['interaction1'] = nsw_dw_cpscontrol.educ*nsw_dw_cpscontrol.re74
# Criando uma variável de interação entre a dummy 'hisp' e a taxa de desemprego de 1974
nsw_dw_cpscontrol['interaction2'] = nsw_dw_cpscontrol.hisp*nsw_dw_cpscontrol.u74
nsw_dw_cpscontrol['re74sq'] = np.power((nsw_dw_cpscontrol.re74), 2) # Criando uma variável da renda de 1974 ao quadrado
nsw_dw_cpscontrol['re75sq'] = np.power((nsw_dw_cpscontrol.re75), 2) # Criando uma variável da renda de 1975 ao quadrado
nsw_dw_cpscontrol.reset_index(inplace=True, drop=True)
nsw_dw_cpscontrol
```

```python
# Estimando o PSM
logit_nsw = smf.glm(formula="""treat ~ age + agesq + agecube + educ + educsq + 
                    marr + nodegree + black + hisp + re74 + re75 + u74 + u75 + interaction1""", 
                    family=sm.families.Binomial(),
                   data=nsw_dw_cpscontrol).fit()
nsw_dw_cpscontrol['pscore'] = logit_nsw.predict(nsw_dw_cpscontrol)
nsw_dw_cpscontrol.groupby('treat')['pscore'].mean()
```

Temos então um valor médio do *pscore* para o grupo de tratamento de 0.43 aproximadamente, enquanto a média para o grupo de controle CPS é de 0.007, arredondando.

```python
# Por fim, plotando as distribuição do 'pscore' para os status de tratamento e controle
p.ggplot(nsw_dw_cpscontrol, p.aes(x='pscore')) +    p.geom_histogram(bins=50) +    p.facet_wrap("treat", scales='free')
```

Como podemos ver nos histogramas acima, nosso grupo de tratamento experimental é muito diferente do grupo de controle não experimental da base do *CPS*.
