### lmb_1

Exercício de replicação do artigo "*Do voters affect or elect policies? Evidence from the U.S. House*" (2004) de Lee, Moretti e Butler (LMB), presente no livro "*Causal Inference: The Mixtape*" referente ao capítulo tratando sobre RDD em eleições apertadas.

```python
# Importando os pacotes necessários
import numpy as np 
import pandas as pd 
import statsmodels.api as sm 
import statsmodels.formula.api as smf 
from itertools import combinations 
import plotnine as p
import ssl
from stargazer.stargazer import Stargazer
import rdd
import matplotlib.pyplot as plt
from localreg import *
import warnings
warnings.filterwarnings('ignore')
```

```python
# Criando uma função de importação/leitura dos dados
ssl._create_default_https_context = ssl._create_unverified_context
def read_data(file):
    return pd.read_stata("https://github.com/scunning1975/mixtape/raw/master/" + file)
```

```python
# Criando uma função de regressão linear robusta
def lm_robust(formula, data):
    regression = sm.OLS.from_formula(formula, data = data)
    regression = regression.fit(cov_type="cluster",cov_kwds={"groups":data['id']})
    return regression
```

```python
# Importando a base direto do github do autor usando a função criada acima
lmb_data = read_data("lmb-data.dta")
lmb_data
```

```python
## Tratando a base
# Criando uma coluna com a 'running variable' centralizada
lmb_data['demvoteshare_c'] = lmb_data['demvoteshare'] - 0.5
# Criando uma amostra com apenas observações dentro do intervalo de 0.48 à 0.52 da 'running variable'
lmb_subset = lmb_data[lmb_data.lagdemvoteshare.between(.48, .52)]
lmb_subset
```

```python
# Reproduzindo as estimações originais de LMB 
lm_1 = lm_robust('score ~ lagdemocrat', data = lmb_subset)
lm_2 = lm_robust('score ~ democrat', data = lmb_subset)
lm_3 = lm_robust('democrat ~ lagdemocrat', data = lmb_subset)
results1 = Stargazer([lm_1, lm_2, lm_3])
results1.title('Resultados Originais baseados nas Pontuações ADA - Amostra de Eleições Apertadas')
results1
```

Os resultados acima indicam um efeito causal positivo e estatísticamente significante da vitória eleitoral democrata sobre os *ADA score* do período seguinte e do período corrente (sendo evidência a favor da teoria da divergência) e sobre a vitória eleitoral democrata no período seguinte (sendo evidência a favor do efeito sobre incumbência).

### lmb_2

Estimando as regressões usando a base completa (ou seja, utilizando as observações de todo a distribuição  da *running variable*).

```python
# Reestimando as regressões usando a amostra completa
lm_1 = lm_robust('score ~ lagdemocrat', data = lmb_data)
lm_2 = lm_robust('score ~ democrat', data = lmb_data)
lm_3 = lm_robust('democrat ~ lagdemocrat', data = lmb_data)
results2 = Stargazer([lm_1, lm_2, lm_3])
results2.title('Resultados Originais Baseados nas Pontuações ADA - Amostra Completa')
results2
```

Quando usamos a amostra completa obtemos resultados diferentes dos anteriores, em que o efeito para a pontuação ADA futura aumenta em aproximadamente 10 pontos e o efeito para a pontuação ADA corrente diminui por volta 7 pontos. O efeito sobre incumbência também apresenta um aumento, passando de 0.48 para 0.82. Todos os resultados apresentaram significância estatística.

### lmb_3

Adicionando controles para a *running variable* e utilizando a *running variable* recentralizada nas regressões.

```python
# Eliminando linhas com valores ausentes na 'running variable' recentralizada
lmb_data = lmb_data[~pd.isnull(lmb_data.demvoteshare_c)]
lmb_data
```

```python
# Reestimando nossas regressões com os ajustes mencionados acima
lm_1 = lm_robust('score ~ lagdemocrat + demvoteshare_c', data = lmb_data)
lm_2 = lm_robust('score ~ democrat + demvoteshare_c', data = lmb_data)
lm_3 = lm_robust('democrat ~ lagdemocrat + demvoteshare_c', data = lmb_data)
results3 = Stargazer([lm_1, lm_2, lm_3])
results3.title('Resultados Originais Baseados nas Pontuações ADA - Amostra Completa')
results3
```

### lmb_4

Para permitir que nossa *running variable* recentralizada varie em ambos os lados da descontinuidade, vamos interagi-lá com nossa variável de tratamento (a parcela de votos do partido Democrata).

```python
# Reestimando nossas regressões interagindo a 'running variable' com a variável de tratamento
lm_1 = lm_robust('score ~ lagdemocrat*demvoteshare_c', data = lmb_data)
lm_2 = lm_robust('score ~ democrat*demvoteshare_c', data = lmb_data)
lm_3 = lm_robust('democrat ~ lagdemocrat*demvoteshare_c', data = lmb_data)
results4 = Stargazer([lm_1, lm_2, lm_3])
results4.title('Resultados Originais Baseados nas Pontuações ADA - Amostra Completa com Interações Lineares')
results4
```

### lmb_5

E como último passo na estimação das regressões globais, vamos adicionar a forma quadrática de nossa *running variable* de modo a lidar com possíveis problemas de não-linearidade.

```python
# Gerando uma coluna com nossa 'running variable' recentralizada quadrada
lmb_data['demvoteshare_sq'] = lmb_data['demvoteshare_c']**2
lmb_data
```

```python
# Reestimando nossas regressões controlando para não-linearidade
lm_1 = lm_robust('score ~ lagdemocrat*demvoteshare_c + lagdemocrat*demvoteshare_sq', 
                 data = lmb_data)
lm_2 = lm_robust('score ~ democrat*demvoteshare_c + democrat*demvoteshare_sq', 
                 data = lmb_data)
lm_3 = lm_robust('democrat ~ lagdemocrat*demvoteshare_c + lagdemocrat*demvoteshare_sq', 
                 data = lmb_data)
results5 = Stargazer([lm_1, lm_2, lm_3])
results5.title('Resultados Originais Baseados nas Pontuações ADA - Amostra Completa com Interações Lineares e Quadráticas')
results5
```

### lmb_6

Restringindo novamente nossa base para estimar o *LATE*.

```python
# Restringindo nosso df global para observações dentro da banda [0.45, 0.55] de nossa 'running variable'
lmb_subset2 = lmb_data[lmb_data.demvoteshare.between(.45, .55)]
lmb_subset2
```

```python
# Reestimando o LATE para uma subamostra com banda de 0.05
lm_1 = lm_robust('score ~ lagdemocrat*demvoteshare_c + lagdemocrat*demvoteshare_sq', 
                 data = lmb_subset2)
lm_2 = lm_robust('score ~ democrat*demvoteshare_c + democrat*demvoteshare_sq', 
                 data = lmb_subset2)
lm_3 = lm_robust('democrat ~ lagdemocrat*demvoteshare_c + lagdemocrat*demvoteshare_sq', 
                 data = lmb_subset2)
results6 = Stargazer([lm_1, lm_2, lm_3])
results6.title('Resultados Baseados nas Pontuações ADA - Amostra Próxima com Interações Lineares e Quadráticas')
results6
```

### lmb_7

Representando gráficamente a relação entre a *running variable* e a pontuação ADA.

```python
# Tratando a base para gerar o gráfico
categories = lmb_subset2.lagdemvoteshare
lmb_subset2['lagdemvoteshare_100'] = pd.cut(lmb_subset2.lagdemvoteshare, 100)
agg_lmb_subset2 = lmb_subset2.groupby('lagdemvoteshare_100')['score'].mean().reset_index()
lmb_subset2['gg_group'] = [1 if x>.5 else 0 for x in lmb_subset2.lagdemvoteshare]
agg_lmb_subset2['lagdemvoteshare'] = np.arange(0.01, 1.01, .01)
agg_lmb_subset2
```

```python
# Plotando o gráfico usando um modelo quadrático para representar a relação entre a 'running variable' e o 'ADA score' 
p.ggplot(lmb_subset2, p.aes('lagdemvoteshare', 'score')) +    p.geom_point(p.aes(x = 'lagdemvoteshare', y = 'score'), data = agg_lmb_subset2) +    p.stat_smooth(p.aes('lagdemvoteshare', 'score', group = 'gg_group'), 
                  data=lmb_subset2, method = "lm", 
              formula = 'y ~ x + I(x**2)') +\
    p.xlim(0,1) + p.ylim(0,100) +\
    p.geom_vline(xintercept = 0.5)
```

```python
# Plotando o gráfico usando um modelo linear para representar a relação entre a 'running variable' e o 'ADA score'
p.ggplot(lmb_subset2, p.aes('lagdemvoteshare', 'score')) +    p.geom_point(p.aes(x = 'lagdemvoteshare', y = 'score'), data = agg_lmb_subset2) +    p.stat_smooth(p.aes('lagdemvoteshare', 'score', group = 'gg_group'), 
                  data=lmb_subset2, method = "lm")+\
    p.xlim(0,1) + p.ylim(0,100) +\
    p.geom_vline(xintercept = 0.5)
```

```python
# Plotando o gráfico usando um ajuste 'lowess' para representar a relação entre a 'running variable' e o 'ADA score'
p.ggplot(lmb_subset2, p.aes('lagdemvoteshare', 'score')) +    p.geom_point(p.aes(x = 'lagdemvoteshare', y = 'score'), data = agg_lmb_subset2) +    p.stat_smooth(p.aes('lagdemvoteshare', 'score', group = 'gg_group'), 
                  data=lmb_subset2, method = "lowess") +\
    p.xlim(0,1) + p.ylim(0,100) +\
    p.geom_vline(xintercept = 0.5)
```
